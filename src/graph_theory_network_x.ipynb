1. Create undirected graph

```python
import networkx as nx
import matplotlib.pyplot as plt

# 定义无向图对象
G = nx.Graph()

# 添加节点
G.add_node(1)  # 添加单个节点1
G.add_nodes_from([
    (2, {"color": "red"}),     # 添加节点2，并设置其颜色属性为红色
    (3, {"color": "green"}),   # 添加节点3，并设置其颜色属性为绿色
    (4, {"color": "blue"}),    # 添加节点4，并设置其颜色属性为蓝色
    (5, {"color": "yellow"})   # 添加节点5，并设置其颜色属性为黄色
])

# 添加边
G.add_edge(1, 2)  # 添加边连接节点1和节点2
G.add_edge(2, 3)  # 添加边连接节点2和节点3
G.add_edge(2, 4)  # 添加边连接节点2和节点4
G.add_edge(3, 5)  # 添加边连接节点3和节点5

# 获取节点的颜色属性
node_colors = [G.nodes[node].get("color", "gray") for node in G.nodes()] #if the color is undefined, use gray color

# 绘制图形
nx.draw(G, with_labels=True, font_size=10, font_family='sans-serif', node_color=node_colors)
plt.show()
```

2. Create Directed Graph

```python
import networkx as nx
import matplotlib.pyplot as plt

# 定义有向图对象
G = nx.DiGraph()

# 添加带颜色属性的节点
G.add_nodes_from([
    (1, {"color": "grey"}),     # 添加节点1，颜色为灰色
    (2, {"color": "red"}),      # 添加节点2，颜色为红色
    (3, {"color": "green"}),    # 添加节点3，颜色为绿色
    (4, {"color": "blue"}),     # 添加节点4，颜色为蓝色
    (5, {"color": "yellow"})    # 添加节点5，颜色为黄色
])

# 添加边
G.add_edge(1, 2)  # 添加从节点1到节点2的边
G.add_edge(2, 3)  # 添加从节点2到节点3的边
G.add_edge(2, 4)  # 添加从节点2到节点4的边
G.add_edge(3, 5)  # 添加从节点3到节点5的边

# 获取图的节点和边的信息
print("Number of nodes:", G.number_of_nodes())  # 打印节点数量
print("Nodes:", G.nodes())                      # 打印所有节点
print("Number of edges:", G.number_of_edges())  # 打印边的数量
print("Edges:", G.edges())                      # 打印所有边

# 提取节点颜色
node_colors = [G.nodes[node]["color"] for node in G.nodes()]  # 获取每个节点的颜色属性

# 绘制图形并显示节点颜色
nx.draw(G, with_labels=True, font_size=15, font_family='sans-serif', node_color=node_colors, cmap=plt.cm.Blues, node_size=500)
plt.show()
```

3. Create Weighted Graph

这个表格详细列出了NetworkX中可用的布局选项（`nx.layout`），用于图的可视化时确定节点的位置。下面是对每种布局的详细解释：

### 1. **bipartite_layout**
   - **说明**：用于绘制二分图，将节点分成两条直线排列。
   - **应用场景**：适合于具有二分结构的图，即图中的节点可以分成两组，组内节点之间无连接，所有边都在两组之间。
   - **主要参数**：
     - `G`：图对象。
     - `nodes`：指定在某一行或列中的节点集合。
     - `align`：对齐方式（水平或垂直排列）。
     - `scale`：控制布局的缩放比例。

### 2. **circular_layout**
   - **说明**：将所有节点放置在一个圆形上。
   - **应用场景**：适用于具有环形结构或无层次关系的图。
   - **主要参数**：
     - `G`：图对象。
     - `scale`：控制圆的大小。
     - `center`：设置圆的中心位置。
     - `dim`：设置圆的维度（一般为2D）。

### 3. **kamada_kawai_layout**
   - **说明**：使用Kamada-Kawai路径长度代价函数定位节点。
   - **应用场景**：适合于需要节点间距离反映图结构的图。
   - **主要参数**：
     - `G`：图对象。
     - `dist`：节点间理想距离。
     - `pos`：初始位置。
     - `weight`：用于边的权重。

### 4. **planar_layout**
   - **说明**：尝试在不相交边的情况下定位节点。
   - **应用场景**：用于平面图（没有边交叉的图）。
   - **主要参数**：
     - `G`：图对象。
     - `scale`：控制布局的缩放比例。
     - `center`：设置布局的中心位置。
     - `dim`：设置布局的维度（一般为2D）。

### 5. **random_layout**
   - **说明**：在单位方块内随机放置节点。
   - **应用场景**：适合用于简单的测试或无特殊布局要求的图。
   - **主要参数**：
     - `G`：图对象。
     - `center`：设置随机布局的中心位置。
     - `dim`：设置布局的维度（一般为2D）。
     - `seed`：随机种子以保证重复实验的一致性。

### 6. **rescale_layout**
   - **说明**：返回缩放到 `(-scale, scale)` 范围内的节点位置数组。
   - **应用场景**：用于调整现有布局的大小。
   - **主要参数**：
     - `pos`：输入的节点位置数组。
     - `scale`：缩放比例。

### 7. **rescale_layout_dict**
   - **说明**：返回以节点为键的缩放后位置的字典。
   - **应用场景**：类似于 `rescale_layout`，但输出为字典格式，方便进一步处理。
   - **主要参数**：
     - `pos`：节点位置的字典。
     - `scale`：缩放比例。

### 8. **shell_layout**
   - **说明**：将节点放置在同心圆上。
   - **应用场景**：适合具有分层结构的图。
   - **主要参数**：
     - `G`：图对象。
     - `nlist`：节点的分层列表，控制节点放置的层。
     - `rotate`：旋转角度，控制圆的起始位置。
     - `scale`：控制圆的大小。

### 9. **spring_layout**
   - **说明**：使用Fruchterman-Reingold力导向算法定位节点。
   - **应用场景**：适用于大多数图，能够较好地展示节点之间的关系。
   - **主要参数**：
     - `G`：图对象。
     - `k`：弹簧的理想距离。
     - `pos`：初始位置。
     - `fixed`：固定某些节点的位置。

### 10. **spectral_layout**
   - **说明**：使用图Laplacian的特征向量来定位节点。
   - **应用场景**：适用于需要节点分布反映图结构的场景。
   - **主要参数**：
     - `G`：图对象。
     - `weight`：边的权重。
     - `scale`：控制布局的缩放比例。
     - `center`：设置布局的中心位置。
     - `dim`：设置布局的维度。

### 11. **spiral_layout**
   - **说明**：将节点放置在螺旋结构中。
   - **应用场景**：适合展示递增关系或顺序关系的图。
   - **主要参数**：
     - `G`：图对象。
     - `scale`：控制螺旋的大小。
     - `center`：设置螺旋的中心位置。
     - `dim`：设置螺旋的维度。

### 12. **multipartite_layout**
   - **说明**：将节点放置在分层直线上。
   - **应用场景**：适用于具有多层结构的图（例如多层网络）。
   - **主要参数**：
     - `G`：图对象。
     - `subset_key`：节点分层的键，控制节点的分层。
     - `align`：对齐方式（垂直或水平）。

---

每种布局方法提供了不同的节点位置方式，根据图的结构和显示需求选择合适的布局，有助于更直观地展示节点和边的关系。可以在 NetworkX 官方文档中找到更多详细信息：[NetworkX layout documentation](https://networkx.github.io/documentation/stable/reference/drawing.html?highlight=layout#module-networkx.drawing.layout)。


```python
# 加权图示例

import matplotlib.pyplot as plt
import networkx as nx

# 创建无向图对象
G = nx.Graph()

# 添加带权重的边
G.add_edge('a', 'b', weight=0.6)  # 添加边 'a' - 'b'，权重为 0.6
G.add_edge('a', 'c', weight=0.2)  # 添加边 'a' - 'c'，权重为 0.2
G.add_edge('c', 'd', weight=0.1)  # 添加边 'c' - 'd'，权重为 0.1
G.add_edge('c', 'e', weight=0.7)  # 添加边 'c' - 'e'，权重为 0.7
G.add_edge('c', 'f', weight=0.9)  # 添加边 'c' - 'f'，权重为 0.9
G.add_edge('a', 'd', weight=0.3)  # 添加边 'a' - 'd'，权重为 0.3

# 根据权重将边分类
elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] > 0.5]  # 权重大于 0.5 的边
esmall = [(u, v) for (u, v, d) in G.edges(data=True) if d['weight'] < 0.5]  # 权重小于 0.5 的边

# 打印边的信息（包括权重）
print(G.edges(data=True))

# 为所有节点生成布局位置
pos = nx.spring_layout(G)    #### has many layout can be used in this graph

# 绘制节点
nx.draw_networkx_nodes(G, pos, node_size=700)

# 绘制边
nx.draw_networkx_edges(G, pos, edgelist=elarge, width=6)  # 绘制权重较大的边
nx.draw_networkx_edges(G, pos, edgelist=esmall, width=6, alpha=0.5, edge_color='b', style='dashed')  # 绘制权重较小的边，使用蓝色虚线

# 添加节点标签
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')

# 关闭坐标轴显示
plt.axis('off')
plt.show()
```

4. calculate centrality measures --- degree centrality for undirected Graph 计算

```python
# 中心性度量示例

import matplotlib.pyplot as plt
import networkx as nx

# 创建一个空手道俱乐部图（Zachary's Karate Club graph）
'''
Zachary's Karate Club 图（Zachary's Karate Club graph）并不是模拟数据，而是基于真实的观察数据构建的。

该图是由社会学家 Wayne W. Zachary 在 1970 年代对一个空手道俱乐部的成员互动进行的实地研究记录。Zachary 观察了俱乐部成员之间的社交关系和互动频率，将每个成员表示为一个节点，成员之间的社交关系表示为边。这张图反映了俱乐部成员的社交网络结构。

后来，由于俱乐部内部发生了争执，俱乐部分裂为两派。Zachary 的研究结果准确预测了成员在分裂后的派别，这一结果广泛用于社交网络分析和聚类算法的测试，是网络科学研究中的经典数据集。

因此，Zachary's Karate Club 图是一组基于真实事件和观察得来的数据，而非模拟数据。
'''
G = nx.karate_club_graph()

# 设置绘图大小并绘制图形
plt.figure(figsize=(15, 15))  # 设置图形尺寸为15x15
nx.draw_networkx(G, with_labels=True)  # 绘制图形并显示节点标签

# 计算度中心性
deg_centrality = nx.degree_centrality(G)  # 使用 NetworkX 的 degree_centrality 函数计算每个节点的度中心性
print("Degree Centrality: " + str(deg_centrality) + '\n')  # 输出每个节点的度中心性


sorted_deg_centrality = sorted(deg_centrality.items(), key=lambda x: x[1], reverse=True) #####排序
# 打印排序后的结果
print("Nodes sorted by Degree Centrality:")
for node, centrality in sorted_deg_centrality:
    print(f"Node {node}: Degree Centrality = {centrality}")
```

6. degree_centrality, closeness_centrality, betweenness_centrality

```python
# 有向图 - 中心性度量示例

import matplotlib.pyplot as plt
import networkx as nx

# 创建有向图对象
G = nx.DiGraph()

# 添加带方向的边
G.add_edge('a', 'b')  # 添加从节点 'a' 到节点 'b' 的有向边
G.add_edge('a', 'c')  # 添加从节点 'a' 到节点 'c' 的有向边
G.add_edge('c', 'd')  # 添加从节点 'c' 到节点 'd' 的有向边
G.add_edge('c', 'e')  # 添加从节点 'c' 到节点 'e' 的有向边
G.add_edge('c', 'f')  # 添加从节点 'c' 到节点 'f' 的有向边
G.add_edge('a', 'd')  # 添加从节点 'a' 到节点 'd' 的有向边

# 生成节点位置，使用 spring 布局
pos = nx.spring_layout(G)

# 定义节点和边的绘制属性
nodes = nx.draw_networkx_nodes(G, pos, alpha=0.5, node_size=700, node_color='blue')  # 绘制节点，设置透明度、大小和颜色
edges = nx.draw_networkx_edges(G, pos, arrowsize=5, edge_color='green', width=0.5)   # 绘制边，设置箭头大小、颜色和宽度

# 添加节点标签
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')  # 设置标签字体大小和字体样式

# 计算度中心性
deg_centrality = nx.degree_centrality(G)
print("Degree Centrality: " + str(deg_centrality))  # 打印每个节点的度中心性

# 计算接近中心性
close_centrality = nx.closeness_centrality(G)
print("Closeness Centrality: " + str(close_centrality))  # 打印每个节点的接近中心性

# 计算介数中心性
between_centrality = nx.betweenness_centrality(G, normalized=True)
print("Betweenness Centrality: " + str(between_centrality))  # 打印每个节点的介数中心性

# 获取当前的绘图轴，并关闭坐标轴显示
ax = plt.gca()
ax.set_axis_off()  # 关闭坐标轴

# 显示绘制的图形
plt.show()
```

8. non-directed grapg clustering

聚类系数（Clustering Coefficient）  - Slide 65-71
-	定义 ：度量节点的邻居是否倾向于形成团体（即邻居间的相互连接情况）。
-	计算 ：包括本地聚类系数和平均聚类系数，用于分析网络的集群特性。
-	主要用于衡量无向图中节点的聚集程度，即一个节点的邻居是否彼此连接。然而，在有向图中也可以计算聚类系数，但其定义和解释会有所不同


```python
# 无向图 - 聚类系数示例

import matplotlib.pyplot as plt
import networkx as nx

# 创建无向图对象
G = nx.Graph()

# 添加边到图中
G.add_edge('a', 'b')  # 添加边 'a' - 'b'
G.add_edge('a', 'c')  # 添加边 'a' - 'c'
G.add_edge('a', 'e')  # 添加边 'a' - 'e'
G.add_edge('c', 'd')  # 添加边 'c' - 'd'
G.add_edge('c', 'e')  # 添加边 'c' - 'e'
G.add_edge('c', 'f')  # 添加边 'c' - 'f'
G.add_edge('a', 'd')  # 添加边 'a' - 'd'
G.add_edge('d', 'e')  # 添加边 'd' - 'e'

# 计算聚类系数
local_clustering_node = nx.clustering(G)  # 计算每个节点的局部聚类系数，返回字典
local_clustering_node_a = nx.clustering(G, 'a')  # 计算节点 'a' 的局部聚类系数
print("Local Clustering: " + str(local_clustering_node))  # 打印所有节点的局部聚类系数
print("Local Clustering for node a: " + str(local_clustering_node_a))  # 打印节点 'a' 的局部聚类系数

# 计算平均聚类系数
avg_clustering = nx.average_clustering(G)  # 计算图的平均聚类系数
print("Average Clustering: " + str(avg_clustering))  # 打印平均聚类系数

# 生成节点位置，使用 spring 布局
pos = nx.spring_layout(G)

# 定义节点和边的绘制属性
nodes = nx.draw_networkx_nodes(G, pos, alpha=0.5, node_size=700, node_color='blue')  # 绘制节点，设置透明度、大小和颜色
edges = nx.draw_networkx_edges(G, pos, edge_color='green', width=0.5)  # 绘制边，设置颜色和宽度

# 添加节点标签
nx.draw_networkx_labels(G, pos, font_size=20, font_family='sans-serif')  # 设置标签字体大小和样式

# 获取当前的绘图轴，并关闭坐标轴显示
ax = plt.gca()
ax.set_axis_off()  # 关闭坐标轴

# 显示绘制的图形
plt.show()
```
